```python
# encoding the target values
encode = {
    "M":0,
    "B":1
}
```

```python
breast["labeled_diagnosis"] = breast["diagnosis"].map(encode)
```

```python
breast.head()
```

```python
breast.drop("diagnosis",axis=1,inplace=True)
```

```python
from sklearn.svm import SVC
```

```python
x = breast.drop(["id","labeled_diagnosis"],axis=1)
y=breast["labeled_diagnosis"]
```

```python
x
```

```python
y
```

```python
# splitting of Data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# model selection
svm = SVC(kernel='linear',random_state=5)
```

```python
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
pred = svm.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
new = pd.DataFrame({
    "actual":y_test,
    "Pred":pred
})
```
