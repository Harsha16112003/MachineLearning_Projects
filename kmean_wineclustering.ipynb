# t0 find the WCSS value of different number of clusters

```python
wcss =[]

for i in range(1,11):
    kmn = KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmn.fit(x)
    wcss.append(kmn.inertia_)
```

```python
# to plot the elbow values 
```

```python
plt.plot(range(1,11),wcss)
plt.title("the elbow points graph")
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
# training the k-means clustering model
kmeans = KMeans(n_clusters=4,init="k-means++",random_state=0)
y = kmeans.fit_predict(x)
y
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c="green",label="c1")
plt.scatter(x[y==1,0],x[y==1,1],s=50,c="blue",label="c2")
plt.scatter(x[y==2,0],x[y==2,1],s=50,c="yellow",label="c3")
plt.scatter(x[y==3,0],x[y==3,1],s=50,c="red",label="c4")
#to draw the centroids
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label="centroids")
plt.show()
```
