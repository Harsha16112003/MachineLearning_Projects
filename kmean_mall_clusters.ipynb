* rename the the column names

```python
mall.rename({"Genre":"gender","Annual Income (k$)":"Annual Income","Spending Score (1-100)":"Spending Score"},inplace=True,axis=1)
```

```python
mall.head()
```

```python
mall.describe()
```

```python
mall.drop("CustomerID",axis=1,inplace=True)
```

```python
mall.head()
```

* extract the data from the mall dataset

```python
x = mall.iloc[:,[2,3]].values
```

```python
x
```

* preprocessing of data

```python
scaler = MinMaxScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

* to find the WCSS value

```python
wcss=[]

for i in range(1,11):
    knn = KMeans(n_clusters=i,init="k-means++",random_state=40)
    knn.fit(x)
    wcss.append(knn.inertia_)
```

* to visualize the elbow graph

```python
plt.plot(range(1,11),wcss)
plt.show()
```

# to train the kmeans clustering model

```python
kmean = KMeans(n_clusters=5,init="k-means++",random_state=4)
y=kmean.fit_predict(x)
y
```

* visualaizing the clusters

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c="green",label="cluster1")
plt.scatter(x[y==1,0],x[y==1,1],s=50,c="red",label="cluster2")
plt.scatter(x[y==2,0],x[y==2,1],s=50,c="blue",label="cluster3")
plt.scatter(x[y==3,0],x[y==3,1],s=50,c="yellow",label="cluster4")
plt.scatter(x[y==4,0],x[y==4,1],s=50,c="pink",label="cluster5")
# to visualize the centroids
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],s=100,c="cyan",marker="*",label="centroids")
plt.title("anual_income VS spending_score")
plt.xlabel("anual_income")
plt.ylabel("spending_score")
plt.legend()
plt.show()
```
